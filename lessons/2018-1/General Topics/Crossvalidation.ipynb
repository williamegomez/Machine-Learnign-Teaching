```python
import numpy as np

def Kfold(y, nfolds = 10):
    uniquetags = np.unique(y)
    indices = [np.empty(0,)] * nfolds
    globaladd = 0
    for j in range(len(uniquetags)):
        inds = np.where(uniquetags[j]==y)[0]
        count = np.sum(uniquetags[j]==y)
        value_per_fold = int(count/nfolds)

        indrandom = np.random.permutation(count)
        inds = inds[indrandom]

        for i in range(nfolds):
            indices[i] = np.concatenate((indices[i], inds[i*value_per_fold:(i+1)*value_per_fold]), axis=0)
        
        lack = count - value_per_fold*nfolds

        indlack = value_per_fold*nfolds
        for i in range(lack):
            indices[globaladd] = np.append(indices[globaladd], inds[indlack+i])
            globaladd = globaladd + 1

            if globaladd == 5:
                globaladd = 0

    output = []  
    for i in range(nfolds):
        output.append([ np.concatenate((np.array(indices)[np.where(np.arange(nfolds) != i)[0]])).astype(int) ,indices[i].astype(int)])

    return output
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()


X = cancer['data']
y = cancer['target']
```

# Validación cruzada
## Se entrena con los demás folds, la estandarización minmax o zscore se hace con el conjunto de entrenamiento, luego los parametros obtenidos se usan para estandarizar los datos de prueba


```python
from sklearn.preprocessing import StandardScaler  
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score

F1ScoreEntrenamiento = np.empty(shape=(0,2))
F1ScorePrueba = np.empty(shape=(0,2))
ErrorEntrenamiento = np.empty(shape=(0,))
ErrorPrueba = np.empty(shape=(0,))

for train_index, test_index in Kfold(y,10):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    #Estandarizacion
    scaler = StandardScaler() 
    scaler.fit(X_train)  # Calculo de parametros
    
    X_train = scaler.transform(X_train)  #Se estandarizan los datos de prueba
    
    
    mlp = MLPClassifier(hidden_layer_sizes=(30,30,30)) 
    mlp.fit(X_train,y_train) #Entrenamiento de la red
    predictions = mlp.predict(X_train) #Predicción con los datos de entrenamiento
    
    ErrorEntrenamiento = np.append(ErrorEntrenamiento,np.sum(predictions!=y_train)/y_train.shape[0])
    F1ScoreEntrenamiento = np.concatenate((F1ScoreEntrenamiento,f1_score(y_train, predictions, average=None)[None]),axis=0)
    
    X_test = scaler.transform(X_test) #Estandarización
    predictions = mlp.predict(X_test) #Predicción
    
    ErrorPrueba = np.append(ErrorPrueba,np.sum(predictions!=y_test)/y_test.shape[0])
    F1ScorePrueba = np.concatenate((F1ScorePrueba,f1_score(y_test, predictions, average=None)[None]),axis=0)
```
